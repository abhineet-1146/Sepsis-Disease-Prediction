###  Data Visulaization for PatientID p001519

```python
df = df.iloc[: , :-7]
df
```

### Visualizing Times series Data

```python
plt.rcParams["figure.figsize"] = (12,30)
df.plot(subplots=True, legend=True)
plt.show()
```

### Histograms

```python
plt.rcParams["figure.figsize"] = (16,18)
fig, axs = plt.subplots(7, 5)
fig.tight_layout()
for _,u in enumerate(df.columns):
    plt.subplot(7,5,_+1)
    df[u].plot(kind='hist')
    plt.title(u)
plt.show()    
```

As we can see in the histogram that at a particular level all the features having frequency at same level and then it decreased drastically
